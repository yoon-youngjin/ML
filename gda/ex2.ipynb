```
import numpy as np
import torch

# 데이터 읽기 함수
def load_dataset(file) :
  data = np.loadtxt(file)
  x = data[:,0:-1]
  y = data[:,-1]

  x = torch.tensor(x,dtype = torch.int)
  y = torch.tensor(y,dtype = torch.int)

  return (x,y)


# 텐서를 리스트로 바꾸는 함수
def tensor2list(tensor_data) :
  return tensor_data.cpu().detach().numpy().tolist()

x,y = load_dataset("/gdrive/My Drive/colab/gda/train.txt")

# 가중치 초기화
w = torch.randn(1)*3.0
# 학습률
alpha = 0.2

# 학습데이터 수
m = y.size(0)

for epoch in range(50) :
  cost = 0
  for i in range(m) :
    cost = cost + ((w*x[i])-y[i])*x[i]
  cost = 1/m *cost
  w = alpha*cost

  print("\nepoch={0:d},w={1}".format(epoch+1,tensor2list(w)))
```
