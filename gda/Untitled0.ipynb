```python
import numpy as np
import torch

def load_data(file) :
  data = np.loadtxt(file)
  x_data = data[:,0:-1]
  y_data = data[:,-1]

  x_data = torch.tensor(x_data,dtype=torch.int)
  y_data = torch.tensor(y_data,dtype=torch.int)

  return (x_data,y_data)
def tensor2list(input) :
  return input.cpu().detach().numpy().tolist()

x,y = load_data("gdrive/My Drive/colab/gda/train.txt")

w = torch.randn(1)*3.0
m = y.size(0)
alpha = 0.2

for epoch in range(50) :
  cost = 0

  for i in range(m) :
    cost = cost + ((w*x[i])-y[i])*x[i]

  cost = 1/m *cost

  w = w - alpha*cost

  print(tensor2list(w))

print("\nresult\n={0},{1}".format(w,m))
```
