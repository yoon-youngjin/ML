```python
import numpy as np
from hmmlearn import hmm

states = ["Rainy","Sunny"]
n_states = len(states)

features = ["walk","shop","clean"]
n_features = len(features)
start_prob = [0.6,0.4]
# 시작 확률
start_probability = np.array([0.6, 0.4])

# 전이 확률
transition_probability = np.array([
  [0.7, 0.3],
  [0.4, 0.6]
])

# 관측 확률
emission_probability = np.array([
  [0.1, 0.4, 0.5],
  [0.6, 0.3, 0.1]
])

model = hmm.MultinomialHMM(n_components=n_states)
model.startprob_ = start_probability
model.transmat_ = transition_probability
model.emission_ = emission_probability

input = [0,0,2,1]
hmm_input = np.atleast_2d(input)
logprob,sequence = model.decode(hmm_input)

print("Input :",",".join(map(lambda x:features[x],input)))
print("Output :",",".join(map(lambda x:states[x],sequence)))
```
